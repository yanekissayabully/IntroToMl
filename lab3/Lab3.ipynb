```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
Y = np.random.normal(loc=5, scale=2, size=100)

plt.hist(Y, bins=30, alpha=0.7, color='red', edgecolor='black')
plt.title("Y")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```

```python
c_val = np.linspace(0, 10, 200) 
mse_val = np.array([np.mean((Y - c) ** 2) for c in c_val])
```

```python
optimal_c = c_val[np.argmin(mse_val)]
mean_y = np.mean(Y)

print(f"optimal c: {optimal_c:.4f}")
print(f"mean Y: {mean_y:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(c_val, mse_val, label="MSE")
plt.axvline(optimal_c, color='red', linestyle='--', label=f"optimal c = {optimal_c:.2f}")
plt.axvline(mean_y, color='green', linestyle=':', label=f"mean Y = {mean_y:.2f}")
plt.xlabel("c")
plt.ylabel("MSE")
plt.title("mse/c")
plt.legend()
plt.show()
```

the shape of the graph is U-shaped (parabola). the minimum is at the mean point of Y. the further c is from the mean, the greater the error, because MSE = mean square deviation.

in my opinion, Dummy is useful because: firstly, it provides a baseline for comparison. if a complex model performs worse than Dummy, then something is wrong. in problems with strong noise, a simple predictor can be quite good

```python
#BONUS

def train_test_split(x, y, test_size=0.2):
    x = np.array(x)
    y = np.array(y)

    n_samples = len(x)
    n_test = int(n_samples * test_size)

    indices = np.arange(n_samples)
    np.random.shuffle(indices)

    test_idx = indices[:n_test]
    train_idx = indices[n_test:]

    x_train, x_test = x[train_idx], x[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    return x_train, x_test, y_train, y_test
```

```python
X = [[1, 2], [3, 4], [5, 6], [7, 8]]
y = [0, 1, 0, 1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

print("x_train:", X_train)
print("x_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)
```
